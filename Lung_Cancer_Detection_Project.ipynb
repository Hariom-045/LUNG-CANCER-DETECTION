IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/lung-cancer/lung-cancer.data'
```

Loading dataset

```python
df=pd.read_csv(url,header=None,skipinitialspace=True ,na_values='?')
```

```python
df.shape
```

```python
#Here 0th column is target variable and all other columns are independent columns.
df.head()
```

```python
df.isnull().sum()
```

DATA PRE-PROCESSING


```python
df[4].fillna(int(df[4].mean()), inplace=True)
df[38].fillna(int(df[38].mean()), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

splitting the data into dependent and independent variable

```python
target=df[0]
```

```python
del df[0]
```

```python
data =df
```

```python
data.head()
```

```python
data.describe()
```

Splitting the data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
data.shape
```

```python
target.shape
```

```python
x_train, x_test, y_train, y_test=train_test_split(data, target, test_size=(0.2))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

SVM PREDICTIONS

```python
train_pred=svc.predict(x_train)
test_pred=svc.predict(x_test)
```

```python
print(accuracy_score(train_pred, y_train))
print(accuracy_score(test_pred, y_test))
```
